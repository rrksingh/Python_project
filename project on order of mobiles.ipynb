```python
mobile_devices = ['apple','samsung','redmi','oppo','vivo']

accesories = ['charger','bluetooth','earphone','cover','tempered glass']

device_price = {'apple':50000,'samsung':35000,'redmi':15000,'oppo':18000,'vivo':16000}

accesorie_price = {'charger':1200,'bluetooth':1600,'earphone':400,'cover':150,'tempered glass':100}

total_amount = []
final_order = {}
customer_detail = {}

print(' Welcome to my mobile store .')

order_device = True

while order_device:
    print('please choose your a device:')
    print()
    for device in mobile_devices:
        print()
        print(device)
        print()
        
    while True:
        device = input('Which mobile device would you like to order:\n')
        if device in mobile_devices:
            print(f'\nyou have order {device}mobile')
                
            final_order.setdefault(device,[])
                
            break
                
        if device not in mobile_devices:
            print()
            print(f'\nsorry we havent{device}device at the time.')
            print()
                    
    print('this is the list of accesories')
    print()
            
    for accesorie in accesories:
        print()
        print(accesorie)
        
    while True:
        
        extra_accesorie = input('Would you like extra accesorie? Yes or No: ')
        if extra_accesorie == 'Yes' or extra_accesorie == 'yes':
            accesorie = input('Which accesorie would you like to add?: ')
            
            if accesorie in accesories:
                print(f' we have add{accesorie}accesorie.')
                
                final_order[device].append(accesorie)
                
            else:
                print(f'sorry we havent{accesorie}accesorie at the moment')
                
        elif extra_accesorie == 'No' or extra_accesorie == 'no': 
            
            break
            
            
    extra_device = input("Would you like to order an extra device?") 
    if extra_device == 'No' or extra_device == 'no':
        
        for key,value in final_order.items():
            print(f'\nYou have ordered {key} device with {value} accesorie.')
            
        check_order = True
        
        while check_order:
            print()
            order_correct = input('Is this correct? yes or no:')
            
            if order_correct=='Yes' or order_correct=='yes':
                
                check_order=False
                order_device=False
                
            if order_correct=='No' or order_correct=='no':
                
                print(final_order)
                
                add_remove = input ('Would you like to add or remove?')
                    
                if add_remove=='remove' or add_remove=='Remove':
                    remove=input('Which device would you like to remove?')
                    
                    del final_order[remove]
                    
                    print(final_order)
                    
                if add_remove=='Add' or add_remove=='add':
                    check_order=False
                             
                    
                                 
print(final_order)

device_order=[]
accesorie_order=[]


for i in final_order.keys():
    #print(i,end=', ')
    device_order.append(i)    
print()

for k in final_order.keys():
    for j in final_order.get(k):
        #print(j,end=', ')
        accesorie_order.append(j)

for tm in device_order:
    total_amount.append(device_price.get(tm))
    
for acc in accesorie_order:
    total_amount.append(accesorie_price.get(acc))
    
    
print(f'\nYour total order price is Rs. {sum(total_amount)}')

print("Please provide us with your name, address and phone number.")
customer_detail['name']=input('Your name please: ')
customer_detail['street_name']=input('Street name and house number:  ')
customer_detail['postalcode']=input('Postal code and city name:  ')
customer_detail['phonenumber']=input('Your number:  ')

print()
print(f"Thank you for your order {customer_detail['name']}.")
print()
print(f'We wil deiver your order to this address.')
print()
print(customer_detail['street_name'])
print(customer_detail['postalcode'])
print()
print(f"We will contact you on {customer_detail['phonenumber']} if anything comes up.")    
                
                
           
            
                
        
        
    
        
            
```
